## GRÁFICA PRESIÓN HIDROSTÁTICA VS TVD

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tvd=int(input('Ingrese TVD del pozo: '))
den= int(input('Ingrese limite maximo de densidad de lodo: '))
stvd= int(input('Ingrese variacion del TVD: '))
sden=float(input('Ingrese variacion de la densidad: '))
```

```python
def lista_tvd(tvd):
    lista_tvd=[]
    for tvd in range(0,tvd+stvd,stvd):
        lista_tvd.append(tvd)
    return lista_tvd
    
```

```python
ph_df=pd.DataFrame({'TVD':lista_tvd(tvd)})
```

```python
contador=1
while contador <= den:  
    ph_df[f'Ph {contador} LPG']=0.052*ph_df['TVD']*contador
    contador=contador+sden
```

```python
ph_df
```

```python
ax=plt.gca()
contador=1
while contador <= den:
    ph_df.plot(x=f'Ph {contador} LPG',y='TVD',ax=ax,legend=True,label=f'Ph {contador} LPG',xlabel='Ph, psia',ylabel='TVD',grid=True,figsize=(10,20),title='Ph vs TVD')
    contador=contador+sden
ax.xaxis.tick_top()
plt.gca().invert_yaxis()
plt.show()

```
